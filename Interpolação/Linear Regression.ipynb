```python
import numpy as np
import sympy as sy
import pandas as pd
from sympy.abc import x,y
import matplotlib.pyplot as plt
sy.init_printing(pretty_print=True)
```

```python
xi=np.arange(1,8)
yi=np.array([0.5,2.5,2.0,4.0,3.5,6.0,5.5])
```

```python
fig, ax = plt.subplots(figsize=(11,7))
_=ax.scatter(xi, yi)
```

$$ a_{1} = \dfrac{n\sum x_{i}y_{i} - \sum x_{i}\sum y_{i}}{n\sum x_{i}^{2}-(\sum x_{i})^{2}}  $$

```python
def A1(xi,yi):
    n=len(xi)
    return (n*np.sum(xi*yi)-np.sum(xi)*np.sum(yi))/(n*np.sum(xi**2)-np.sum(xi)**2)
```

```python
a1=A1(xi,yi);a1
```

$$ a_{0} = \overline{y} - a_{1}\overline{x}$$

```python
def A0(xi,yi,a1):
    n=len(xi)
    return (np.sum(yi)/n)-a1*(np.sum(xi)/n)
```

```python
a0=A0(xi,yi,a1);a0
```

$$ y = a_{0} + a_{1}x  $$

```python
y=a0+a1*x;y
```

```python
f=sy.lambdify(x,y,'numpy')
x=np.linspace(0,8)
```

```python
fig, ax = plt.subplots(figsize=(11,7))
ax.scatter(xi, yi)
_=plt.plot(x,f(x))
```

$$ r = \dfrac{n\sum x_{i}y_{i}-(\sum x_{i})(\sum y_{i})}{\sqrt{n\sum x_{i}^{2}-(\sum x_{i})^{2}}\sqrt{n\sum y_{i}^{2}-(\sum y_{i})^{2}}} $$

```python
def R(xi,yi):
    n=len(xi)
    aux1=np.sqrt(n*np.sum(xi**2)-(np.sum(xi)**2))
    aux2=np.sqrt(n*np.sum(yi**2)-(np.sum(yi)**2))
    return (n*np.sum(xi*yi)-(np.sum(xi)*np.sum(yi)))/(aux1*aux2)
```

```python
r=R(xi,yi);r
```

```python
r=R(xi,yi);r
```
